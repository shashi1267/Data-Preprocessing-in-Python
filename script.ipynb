```python
# importing libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

# Q1

```python
# reading the data
data=pd.read_csv('Restaurant Reviews.csv')
```

# Q2

```python
# subsetting the month, grouping the data by month and thene agregating and calculating mean , std and no of reviews.
data['Month/Year']=data['date'].apply(lambda x:'-'.join(x.split('-')[0:2]))
reviews_by_date=data.groupby('Month/Year').agg({'rating':['count','mean','std']})
reviews_by_date=reviews_by_date.reset_index()
reviews_by_date.columns=['Month/Year','Number of Reviews','Average','Std. Dev.']
reviews_by_date['Std. Dev.']=reviews_by_date['Std. Dev.'].fillna(0)
reviews_by_date.head()
```

# Q3

```python
# grouping the data by restaurant and then agregating and calculating mean , std and no of reviews.FInally sorting 
#by Number of revies and showing top 20

most_reviewed=data.groupby('restaurant').agg({'rating':['count','mean','std']})
most_reviewed=most_reviewed.reset_index()
most_reviewed.columns=['Restaurant Name','Number of Reviews','Average','Std. Dev.']
most_reviewed=most_reviewed.sort_values(by='Number of Reviews',ascending=False)
most_reviewed.head(20)
```

# Q4

```python
# grouping the data by restaurant and then agregating and calculating mean , std and no of reviews.FInally sorting 
#by average of ratings and showing top 20

highest_avg=data.groupby('restaurant').agg({'rating':['count','mean','std']})
highest_avg=highest_avg.reset_index()
highest_avg.columns=['Restaurant Name','Number of Reviews','Average','Std. Dev.']
highest_avg=highest_avg.sort_values(by='Average',ascending=False)
highest_avg.head(20)
```

# Q5

```python
# plotting graph for timeseries data using matplotlib
x_axis=reviews_by_date.head(20)['Month/Year']
data1=reviews_by_date.head(20)['Number of Reviews']
data2=reviews_by_date.head(20)['Average']
data3=reviews_by_date.head(20)['Std. Dev.']
line1 = plt.plot(x_axis, data1,'ko-',label='Number Of Reviews')
line2 = plt.plot(x_axis, data2,'ro-',label='Average') 
line3 = plt.plot(x_axis, data3,'mo-',label='Std. Dev.')
plt.xlabel("Month/Year")
plt.ylabel("value")
plt.legend(loc="best")
plt.xticks(rotation=90)
plt.title('time-series data by month')
plt.show()
```

In the timeseries data we have x-axis as year month and y axis as the value.There are three legends in the data. One shows the count of reviews that is nothing but no of reviews given, average that is nothing but average of review and third is standard deviation. Many of the month we have only single review so standard deviation is none for them which we have marked as zero. The data currently doesn’t show any significant insight.
